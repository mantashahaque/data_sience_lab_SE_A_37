### 1. Program to compute the factorial of a given number

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Test the function
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

### 2. Program to display the multiplication table for a given number

```python
def display_multiplication_table(number, limit=10):
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Test the function
display_multiplication_table(7)
```

### 3. Program to identify if a given number is odd or even

```python
def check_odd_even(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Test the function
num1 = 4
print(f"The number {num1} is {check_odd_even(num1)}.")

num2 = 7
print(f"The number {num2} is {check_odd_even(num2)}.")
```

### 4. Program to identify if the entered number is a palindrome or not

```python
def is_palindrome(number):
    # Convert number to string to easily reverse it
    s_number = str(number)
    return s_number == s_number[::-1]

# Test the function
num1 = 121
print(f"Is {num1} a palindrome? {is_palindrome(num1)}")

num2 = 12345
print(f"Is {num2} a palindrome? {is_palindrome(num2)}")

num3 = 1221
print(f"Is {num3} a palindrome? {is_palindrome(num3)}")
```

### 5. Develop an Expert System that provides simple decision-making

Here's a simple rule-based expert system for recommending an activity based on weather and time of day. This system uses a set of 'if-then' rules to make a decision.

```python
class SimpleExpertSystem:
    def __init__(self):
        self.rules = {
            ('sunny', 'day'): 'Go for a walk in the park.',
            ('rainy', 'day'): 'Read a book indoors.',
            ('cloudy', 'day'): 'Visit a museum.',
            ('sunny', 'night'): 'Stargaze.',
            ('rainy', 'night'): 'Watch a movie.',
            ('cloudy', 'night'): 'Play board games.'
        }

    def get_recommendation(self, weather, time_of_day):
        condition = (weather.lower(), time_of_day.lower())
        if condition in self.rules:
            return f"Based on the weather ({weather}) and time ({time_of_day}), I recommend: {self.rules[condition]}"
        else:
            return "I don't have a specific recommendation for these conditions."

# Create an instance of the expert system
es = SimpleExpertSystem()

# Get recommendations
print(es.get_recommendation('Sunny', 'Day'))
print(es.get_recommendation('Rainy', 'Night'))
print(es.get_recommendation('Cloudy', 'Day'))
print(es.get_recommendation('Snowy', 'Day')) # Condition not in rules
```
