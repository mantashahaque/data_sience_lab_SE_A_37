## Implement BFS and DFS

### Subtask:
Implement Breadth-First Search (BFS) and Depth-First Search (DFS) algorithms in Python, define a graph structure, and provide example usage for each.


**Reasoning**:
The subtask requires defining a graph, implementing BFS and DFS algorithms, and then demonstrating their usage. I will start by defining a simple graph structure and then implement the BFS algorithm as specified in the instructions, using a queue and keeping track of visited nodes.



```python
import collections

# 1. Define a graph using an adjacency list representation
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("Graph defined:")
for node, neighbors in graph.items():
    print(f"  {node}: {neighbors}")

# 2. Implement the Breadth-First Search (BFS) algorithm
def bfs(graph, start_node):
    visited = set() # To keep track of visited nodes
    queue = collections.deque([start_node]) # Queue for BFS
    bfs_traversal = [] # To store the order of visited nodes

    visited.add(start_node)

    while queue:
        current_node = queue.popleft() # Dequeue the front node
        bfs_traversal.append(current_node)

        # Enqueue all unvisited neighbors
        for neighbor in graph.get(current_node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return bfs_traversal

print("\nBFS algorithm implemented.")
```

**Reasoning**:
The next step is to implement the Depth-First Search (DFS) algorithm as per the instructions, and then demonstrate the usage of both BFS and DFS algorithms with the defined graph.



```python
import collections

# 3. Implement the Depth-First Search (DFS) algorithm
def dfs(graph, start_node):
    visited = set()  # To keep track of visited nodes
    stack = [start_node]  # Stack for DFS (using a Python list)
    dfs_traversal = []  # To store the order of visited nodes

    while stack:
        current_node = stack.pop()  # Pop the top node
        if current_node not in visited:
            visited.add(current_node)
            dfs_traversal.append(current_node)

            # Push unvisited neighbors onto the stack (in reverse order for consistent output if order matters)
            # graph.get(current_node, [])[::-1] ensures neighbors are added to stack in reverse,
            # so they are popped in the 'natural' order (e.g., A then B if ['B', 'A'] was reversed)
            for neighbor in sorted(graph.get(current_node, []), reverse=True):
                if neighbor not in visited:
                    stack.append(neighbor)
    return dfs_traversal

print("DFS algorithm implemented.\n")

# 5. Call the BFS function with the example graph and a chosen starting node
start_node = 'A'
bfs_result = bfs(graph, start_node)
print(f"BFS Traversal from node {start_node}: {bfs_result}")

# 6. Call the DFS function with the same example graph and starting node
dfs_result = dfs(graph, start_node)
print(f"DFS Traversal from node {start_node}: {dfs_result}")
```

## Implement A* Algorithm

### Subtask:
Implement the A* search algorithm in Python, including defining a graph/grid, a heuristic function, and the A* function to find the shortest path between a start and an end node. Provide an example usage to demonstrate its functionality.


**Reasoning**:
The subtask requires implementing the A* algorithm. I will start by defining a grid (represented by its dimensions), implementing a heuristic function (Manhattan distance for a grid), and setting up the basic A* algorithm structure using a priority queue.



```python
import heapq

# 1. Define a grid or graph structure
# For A* on a grid, we'll represent the grid as a 2D space.
# The 'graph' will be implicitly defined by possible movements (up, down, left, right).
# We'll define obstacles as part of the grid later or within the neighbor generation.

# Let's define the grid dimensions for a conceptual grid
grid_width = 10
grid_height = 10

# Define obstacles as a set of coordinates (row, col)
obstacles = {
    (1, 5), (2, 5), (3, 5), (4, 5),
    (6, 3), (6, 4), (6, 5), (6, 6)
}

print("Grid dimensions defined:", grid_width, "x", grid_height)
print("Obstacles defined:", obstacles)

# 2. Implement a heuristic function (Manhattan distance for a grid)
def heuristic(node, goal):
    """Calculates the Manhattan distance between two nodes."""
    return abs(node[0] - goal[0]) + abs(node[1] - goal[1])

print("Heuristic function (Manhattan distance) implemented.")

# 3. Implement the A* search algorithm
def a_star(start, goal, width, height, obstacles):
    # Priority queue: (f_score, node)
    open_set = []
    heapq.heappush(open_set, (0, start))

    # came_from: dictionary to reconstruct path
    came_from = {}

    # g_score: actual cost from start to current node
    g_score = {node: float('inf') for row in range(height) for node in [(row, col) for col in range(width)]}
    g_score[start] = 0

    # f_score: estimated total cost from start to goal through current node
    f_score = {node: float('inf') for row in range(height) for node in [(row, col) for col in range(width)]}
    f_score[start] = heuristic(start, goal)

    while open_set:
        current_f_score, current_node = heapq.heappop(open_set)

        if current_node == goal:
            # Reconstruct path
            path = []
            temp = current_node
            while temp in came_from:
                path.append(temp)
                temp = came_from[temp]
            path.append(start)
            return path[::-1], g_score[goal]

        # Define neighbors (up, down, left, right)
        row, col = current_node
        neighbors = [
            (row - 1, col), (row + 1, col),
            (row, col - 1), (row, col + 1)
        ]

        for neighbor in neighbors:
            n_row, n_col = neighbor

            # Check if neighbor is within grid bounds and not an obstacle
            if 0 <= n_row < height and 0 <= n_col < width and neighbor not in obstacles:
                tentative_g_score = g_score[current_node] + 1 # Cost to move to an adjacent cell is 1

                if tentative_g_score < g_score[neighbor]:
                    came_from[neighbor] = current_node
                    g_score[neighbor] = tentative_g_score
                    f_score[neighbor] = g_score[neighbor] + heuristic(neighbor, goal)
                    heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return None, float('inf') # Path not found

print("A* search algorithm implemented.")
```

**Reasoning**:
The A* algorithm has been implemented, so the next step is to demonstrate its functionality by defining start and end nodes, calling the A* function, and printing the resulting path and its cost as per the subtask instructions.



```python
print("\n4. Example Usage of A* Algorithm:")

# 4. Provide an example usage: define a start node, an end node
start_node_a_star = (0, 0) # Top-left corner
goal_node_a_star = (9, 9) # Bottom-right corner

print(f"  Start node: {start_node_a_star}")
print(f"  Goal node: {goal_node_a_star}")

# Call the A* function
path, cost = a_star(start_node_a_star, goal_node_a_star, grid_width, grid_height, obstacles)

# 5. Print the found path and its total cost
if path:
    print(f"  Path found by A*: {path}")
    print(f"  Total cost of the path: {cost}")
else:
    print("  No path found by A*.")
```

## Implement Minimax Algorithm

### Subtask:
Implement the basic Minimax algorithm for a two-player deterministic game (e.g., Tic-Tac-Toe) in Python. This step should include defining the game state, possible moves, and the utility function, then writing the Minimax function to determine the optimal move for a given player. Include an example usage to demonstrate its functionality.


**Reasoning**:
I will start by defining the Tic-Tac-Toe board representation and a function to determine all possible moves from a given state, which corresponds to the first two instructions of the subtask.



```python
import math

# Define constants for players and empty cells
EMPTY = ' '
PLAYER_X = 'X'
PLAYER_O = 'O'

# 1. Define the initial game state for Tic-Tac-Toe
# A 3x3 board represented as a list of lists
def create_initial_board():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

# 2. Implement a function to determine all possible next moves from a given game state
def get_possible_moves(board):
    moves = []
    for r in range(3):
        for c in range(3):
            if board[r][c] == EMPTY:
                moves.append((r, c))
    return moves

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("---------")

print("Tic-Tac-Toe board representation and possible moves function implemented.")

# Example of initial board and possible moves
initial_board = create_initial_board()
print("\nInitial Board:")
print_board(initial_board)

print("\nPossible moves from initial state:", get_possible_moves(initial_board))
```

**Reasoning**:
The next step is to implement a function to evaluate the utility (score) of a terminal game state, checking for win conditions and draws, as per the third instruction of the subtask.



```python
import math

# Define constants for players and empty cells (already defined in previous cell, re-included for clarity/completeness)
EMPTY = ' '
PLAYER_X = 'X'
PLAYER_O = 'O'

# Helper function to check for a win
def check_win(board, player):
    # Check rows
    for row in board:
        if all(s == player for s in row):
            return True
    # Check columns
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True
    # Check diagonals
    if all(board[i][i] == player for i in range(3)) or \
       all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

# 3. Implement a function to evaluate the utility (score) of a terminal game state
def evaluate(board):
    if check_win(board, PLAYER_X):
        return 1 # PLAYER_X wins
    elif check_win(board, PLAYER_O):
        return -1 # PLAYER_O wins
    elif not get_possible_moves(board): # No more moves and no winner
        return 0 # Draw
    return None # Game is not over yet

print("Utility evaluation function implemented.")

# Example of a winning state for PLAYER_X
winning_board_x = [
    [PLAYER_X, PLAYER_X, PLAYER_X],
    [PLAYER_O, PLAYER_O, EMPTY],
    [EMPTY, EMPTY, EMPTY]
]
print("\nExample Board (PLAYER_X wins):")
print_board(winning_board_x)
print(f"Evaluation: {evaluate(winning_board_x)}")

# Example of a winning state for PLAYER_O
winning_board_o = [
    [PLAYER_X, PLAYER_X, EMPTY],
    [PLAYER_O, PLAYER_O, PLAYER_O],
    [EMPTY, EMPTY, EMPTY]
]
print("\nExample Board (PLAYER_O wins):")
print_board(winning_board_o)
print(f"Evaluation: {evaluate(winning_board_o)}")

# Example of a draw state
draw_board = [
    [PLAYER_X, PLAYER_O, PLAYER_X],
    [PLAYER_X, PLAYER_O, PLAYER_O],
    [PLAYER_O, PLAYER_X, PLAYER_X]
]
print("\nExample Board (Draw):")
print_board(draw_board)
print(f"Evaluation: {evaluate(draw_board)}")

# Example of a non-terminal state
non_terminal_board = [
    [PLAYER_X, EMPTY, PLAYER_X],
    [PLAYER_O, PLAYER_O, EMPTY],
    [EMPTY, EMPTY, EMPTY]
]
print("\nExample Board (Non-terminal):")
print_board(non_terminal_board)
print(f"Evaluation: {evaluate(non_terminal_board)}")
```

**Reasoning**:
The next step is to implement the Minimax algorithm itself, which should recursively explore game states, assuming optimal play from both players, and return the optimal score for the current player, as per instruction 4 of the subtask.



```python
import math # Re-import for clarity, though already imported

# Define constants for players and empty cells (already defined, re-included for clarity/completeness)
EMPTY = ' '
PLAYER_X = 'X' # Maximizing player
PLAYER_O = 'O' # Minimizing player

# Helper function to apply a move to a board
def apply_move(board, move, player):
    new_board = [row[:] for row in board] # Create a deep copy of the board
    r, c = move
    new_board[r][c] = player
    return new_board

# 4. Implement the Minimax algorithm
def minimax(board, depth, maximizing_player):
    score = evaluate(board)

    # If game is over (terminal state), return the score
    if score is not None:
        return score

    if maximizing_player:
        max_eval = -math.inf
        for move in get_possible_moves(board):
            new_board = apply_move(board, move, PLAYER_X)
            evaluation = minimax(new_board, depth + 1, False)
            max_eval = max(max_eval, evaluation)
        return max_eval
    else:
        min_eval = math.inf
        for move in get_possible_moves(board):
            new_board = apply_move(board, move, PLAYER_O)
            evaluation = minimax(new_board, depth + 1, True)
            min_eval = min(min_eval, evaluation)
        return min_eval

print("Minimax algorithm implemented.")
```

**Reasoning**:
Now that the `minimax` function is implemented, I will create a function to find the best move for the current player by iterating through all possible moves, simulating each move, and using the `minimax` algorithm to evaluate the resulting board state. This addresses instruction 5 of the subtask.



```python
import math # Re-import for clarity

# Define constants for players and empty cells
EMPTY = ' '
PLAYER_X = 'X' # Maximizing player
PLAYER_O = 'O' # Minimizing player

# Assuming create_initial_board, get_possible_moves, check_win, evaluate, apply_move and minimax functions are already defined

# 5. Create a function to find the best move for the current player using the Minimax algorithm
def find_best_move(board, player):
    best_move = None

    if player == PLAYER_X: # Maximizing player
        best_eval = -math.inf
        for move in get_possible_moves(board):
            new_board = apply_move(board, move, player)
            # The minimax function evaluates the opponent's (minimizing player's) perspective
            evaluation = minimax(new_board, 0, False)
            if evaluation > best_eval:
                best_eval = evaluation
                best_move = move
        return best_move, best_eval
    else: # Minimizing player
        best_eval = math.inf
        for move in get_possible_moves(board):
            new_board = apply_move(board, move, player)
            # The minimax function evaluates the opponent's (maximizing player's) perspective
            evaluation = minimax(new_board, 0, True)
            if evaluation < best_eval:
                best_eval = evaluation
                best_move = move
        return best_move, best_eval

print("Function to find the best move implemented.")
```

**Reasoning**:
The previous steps have implemented all necessary functions for the Minimax algorithm, including `find_best_move`. The final instruction of the subtask is to provide an example usage. Therefore, I will define an example game state and use `find_best_move` to demonstrate how the optimal move is determined for the maximizing player, and print the results.



```python
print("\n6. Example Usage of Minimax Algorithm:")

# Example 1: Starting from an empty board, PLAYER_X (maximizing player) makes the first move.
current_board_1 = create_initial_board()
print("\n--- Initial Empty Board (PLAYER_X to move) ---")
print_board(current_board_1)

best_move_1, best_eval_1 = find_best_move(current_board_1, PLAYER_X)
print(f"Optimal move for {PLAYER_X}: {best_move_1} with evaluation: {best_eval_1}")

# Simulate the move and print the new board
new_board_1 = apply_move(current_board_1, best_move_1, PLAYER_X)
print("Board after PLAYER_X's optimal move:")
print_board(new_board_1)

# Example 2: A mid-game state where PLAYER_X needs to find the best move.
current_board_2 = [
    [PLAYER_X, PLAYER_O, EMPTY],
    [PLAYER_X, EMPTY, PLAYER_O],
    [EMPTY, EMPTY, EMPTY]
]
print("\n--- Mid-Game Board (PLAYER_X to move) ---")
print_board(current_board_2)

best_move_2, best_eval_2 = find_best_move(current_board_2, PLAYER_X)
print(f"Optimal move for {PLAYER_X}: {best_move_2} with evaluation: {best_eval_2}")

# Simulate the move and print the new board
new_board_2 = apply_move(current_board_2, best_move_2, PLAYER_X)
print("Board after PLAYER_X's optimal move:")
print_board(new_board_2)
```

## Summary:

### Data Analysis Key Findings

*   **Breadth-First Search (BFS):**
    *   **Purpose:** Explores a graph level by level to find the shortest path in an unweighted graph or to traverse all nodes systematically.
    *   **Implementation:** Utilized a `collections.deque` as a queue and a set for visited nodes.

*   **Depth-First Search (DFS):**
    *   **Purpose:** Explores as far as possible along each branch before backtracking, typically used for pathfinding, topological sorting, or cycle detection.
    *   **Implementation:** Utilized a Python list as a stack and a set for visited nodes.
    *   **Demonstrated Output:** For the same graph, DFS traversal from node 'A' resulted in `['A', 'B', 'D', 'E', 'F', 'C']`, demonstrating its deep exploration before backtracking.
*   **A\* Search:**
    *   **Purpose:** Finds the shortest path between a start and a goal node in a graph, considering both the cost to reach a node (g-score) and the estimated cost to reach the goal from that node (heuristic/h-score).
    *   **Implementation:** Used a 10x10 grid with obstacles, the Manhattan distance as the heuristic function, and `heapq` for priority queue management based on `f_score` (g-score + h-score).

*   **Minimax Algorithm:**
    *   **Purpose:** A decision-making algorithm used in game theory to find the optimal move for a player, assuming the opponent also plays optimally.
    *   **Implementation:** Modeled for a Tic-Tac-Toe game, including functions for game state representation, possible moves, and a utility function to evaluate terminal states (1 for 'X' win, -1 for 'O' win, 0 for draw). The `minimax` function recursively explored game states.
    *   **Demonstrated Output:**
        *   From an initial empty board, the algorithm identified an optimal move for 'X' (e.g., `(0, 0)`) with an evaluation of `0`, indicating a guaranteed draw with optimal play.
